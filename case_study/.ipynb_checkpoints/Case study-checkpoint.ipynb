## Case Study

```python
data = [
    {"app": "A", "bid_price": 0.01, "win": 0, "events": 100000},
    {"app": "A", "bid_price": 0.01, "win": 1, "events": 0},
    {"app": "A", "bid_price": 0.1, "win": 0, "events": 7000},
    {"app": "A", "bid_price": 0.1, "win": 1, "events": 3000},
    {"app": "A", "bid_price": 0.2, "win": 0, "events": 8000000},
    {"app": "A", "bid_price": 0.2, "win": 1, "events": 2000000},
    {"app": "A", "bid_price": 0.4, "win": 0, "events": 700000},
    {"app": "A", "bid_price": 0.4, "win": 1, "events": 300000},
    {"app": "A", "bid_price": 0.5, "win": 0, "events": 80000},
    {"app": "A", "bid_price": 0.5, "win": 1, "events": 20000},
    {"app": "A", "bid_price": 0.75, "win": 0, "events": 7000},
    {"app": "A", "bid_price": 0.75, "win": 1, "events": 3000},
    {"app": "A", "bid_price": 1, "win": 0, "events": 400},
    {"app": "A", "bid_price": 1, "win": 1, "events": 600},
    {"app": "A", "bid_price": 2, "win": 0, "events": 30},
    {"app": "A", "bid_price": 2, "win": 1, "events": 70},
    {"app": "A", "bid_price": 5, "win": 0, "events": 2},
    {"app": "A", "bid_price": 5, "win": 1, "events": 8},
    {"app": "A", "bid_price": 9, "win": 0, "events": 0},
    {"app": "A", "bid_price": 9, "win": 1, "events": 1}
]
```

### Problem1 
Estimate the expected win rate for a bid repsonse at a given price.

1. The win rate is the probability of winning an auction at a given bid price.
2. The win rate can be calculated as the number of wins divided by the total number of events for each bid price.
   #### Win_rate = win_events/(win_events+loss_events) for a particular bid price

```python


df = pd.DataFrame(data)

# Group by bid_price and calculate win rate
win_rate_df = df.groupby('bid_price').apply(lambda x: pd.Series({
    'win_events': x[x['win'] == 1]['events'].sum(),
    'total_events': x['events'].sum(),
    'win_rate': x[x['win'] == 1]['events'].sum() / x['events'].sum()
})).reset_index()

display(win_rate_df)
```

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.plot(win_rate_df['bid_price'], win_rate_df['win_rate'], marker='o')
plt.xlabel('Bid Price')
plt.ylabel('Win Rate')
plt.title('Win Rate vs Bid Price')
plt.grid(True)
plt.show()
```

Win rate increases with bid price, and we see a sharp increase in win_rate at bid price 1.

### Problem 2 
If our goal was to maximize net revenue, what is the most optimal bid valuation we should send in our response?

#### Assumption : 
In order to calculate net revenue we would need to calculate revenue per bid win.It is mentioned that the highest bid price is the amount that the advertiser is willing to pay for bid win.
##### Lets assume that revenue per win = 9 (highest bid from the data)

##### expected_revenue = win_rate * revenue_per_win
##### expected_cost = win_rate * bid_price
##### net_revenue = expected_revenue - expected_cost

```python

def calculate_optimal_bid(win_rate_df, revenue_per_win):
    # Calculate expected revenue, cost, and net revenue
    win_rate_df['expected_revenue'] = win_rate_df['win_rate'] * revenue_per_win
    win_rate_df['expected_cost'] = win_rate_df['bid_price'] * win_rate_df['win_rate']
    win_rate_df['net_revenue'] = win_rate_df['expected_revenue'] - win_rate_df['expected_cost']

    # Find the optimal bid price
    optimal_bid = win_rate_df.loc[win_rate_df['net_revenue'].idxmax()]

    return optimal_bid['bid_price'], optimal_bid['net_revenue']


revenue_per_win = 9
optimal_bid_price, expected_net_revenue = calculate_optimal_bid(win_rate_df, revenue_per_win)

print(f"Optimal bid price: {optimal_bid_price}")
print(f"Expected net revenue: {expected_net_revenue}")

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(win_rate_df['bid_price'], win_rate_df['net_revenue'], marker='o')
plt.xlabel('Bid Price')
plt.ylabel('Expected Net Revenue')
plt.title('Expected Net Revenue vs Bid Price')
plt.grid(True)
plt.annotate(f'Optimal Bid: ${optimal_bid_price:.2f}\nMax Net Revenue: ${expected_net_revenue:.2f}',
             xy=(optimal_bid_price, expected_net_revenue),
             xytext=(optimal_bid_price + 0.5, expected_net_revenue + 0.02),
             arrowprops=dict(facecolor='black', shrink=0.05),
             fontsize=5,
             bbox=dict(facecolor='white', edgecolor='black'))

plt.legend()
plt.show()
```

```python
print(f"Optimal bid price: {optimal_bid_price}")
print(f"Expected net revenue: {expected_net_revenue}")
```

#### Thin data issue
The data for higher bids is very thin, and hence the win_rate calculated on this thin data is not very reliable.Possible solutions
1. Aggregate data across similar bid prices to increase the sample size
2. Use lower bid prices with more data to extrapolate the expected win rates for higher bid prices using linear regression
3. Bayesian Smoothing: Use a Bayesian approach to combine prior knowledge with observed data.
